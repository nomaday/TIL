##### No. 1149 RGB거리
>**문제**<br>
RGB거리에는 집이 N개 있다. 거리는 선분으로 나타낼 수 있고, 1번 집부터 N번 집이 순서대로 있다.<br>
집은 빨강, 초록, 파랑 중 하나의 색으로 칠해야 한다. 각각의 집을 빨강, 초록, 파랑으로 칠하는 비용이 주어졌을 때, 아래 규칙을 만족하면서 모든 집을 칠하는 비용의 최솟값을 구해보자.<br>
1번 집의 색은 2번 집의 색과 같지 않아야 한다.<br>
N번 집의 색은 N-1번 집의 색과 같지 않아야 한다.<br>
i(2 ≤ i ≤ N-1)번 집의 색은 i-1번, i+1번 집의 색과 같지 않아야 한다.<br><br>
**입력**<br>
첫째 줄에 집의 수 N(2 ≤ N ≤ 1,000)이 주어진다. 둘째 줄부터 N개의 줄에는 각 집을 빨강, 초록, 파랑으로 칠하는 비용이 1번 집부터 한 줄에 하나씩 주어진다. 집을 칠하는 비용은 1,000보다 작거나 같은 자연수이다.<br><br>
**출력**<br>
첫째 줄에 모든 집을 칠하는 비용의 최솟값을 출력한다.

```python
# 집 N개     # 거리: 선분    # 집 color: R, G, B 택 1
# 1집색 != 2집색
# N != N-1     # i != i-1       # i != i+1
# R G B 중에 최솟값 찾기
# 최소비용 색상 배정 (앞집하고 색깔 같으면 다음 적은 비용 선택)
# R(3개중 1개)    GB(2개중 1개)      B(2개: 나머지 1개 or 1집 색)

N = int(input())
RGB = []
for _ in range(N):
    RGB.append(list(map(int, input().split())))
#print(RGB)
 
for i in range(1, N):
    # i번째 집을 Red로 했을 때
    RGB[i][0] = RGB[i][0] +    min(RGB[i-1][1], RGB[i-1][2])  # i번째 집의 Red 비용 + 앞집의 Green과 Blue 비용중 최소값
    
    # i번째 집을 Green으로 했을 때
    RGB[i][1] = RGB[i][1] +    min(RGB[i-1][0], RGB[i-1][2])  # i번째 집의 Green 비용 + 앞집의 Red와 Blue 비용중 최소값
        
    # i번째 집을 Blue로 했을 때
    RGB[i][2] = RGB[i][2] +    min(RGB[i-1][0], RGB[i-1][1])  # i번째 집의 Blue 비용 + 앞집의 Red와 Green 비용중 최소값

# 이 값들 중 최소값 구하면 됨    (왜 N-1 이 들어가는지??) => i 가 N의 요소이고 , 집의 갯수가 N개인데 0부터 시작하기때문에 N-1로 넣어줌
print(min(RGB[N-1][0], RGB[N-1][1], RGB[N-1][2]))
```

##### No. 2156 포도주 시식
>**문제**<br>
효주는 포도주 시식회에 갔다. 그 곳에 갔더니, 테이블 위에 다양한 포도주가 들어있는 포도주 잔이 일렬로 놓여 있었다. 효주는 포도주 시식을 하려고 하는데, 여기에는 다음과 같은 두 가지 규칙이 있다.<br>
<br>
포도주 잔을 선택하면 그 잔에 들어있는 포도주는 모두 마셔야 하고, 마신 후에는 원래 위치에 다시 놓아야 한다. 연속으로 놓여 있는 3잔을 모두 마실 수는 없다.<br>
효주는 될 수 있는 대로 많은 양의 포도주를 맛보기 위해서 어떤 포도주 잔을 선택해야 할지 고민하고 있다. 1부터 n까지의 번호가 붙어 있는 n개의 포도주 잔이 순서대로 테이블 위에 놓여 있고, 각 포도주 잔에 들어있는 포도주의 양이 주어졌을 때, 효주를 도와 가장 많은 양의 포도주를 마실 수 있도록 하는 프로그램을 작성하시오. <br>
<br>
예를 들어 6개의 포도주 잔이 있고, 각각의 잔에 순서대로 6, 10, 13, 9, 8, 1 만큼의 포도주가 들어 있을 때, 첫 번째, 두 번째, 네 번째, 다섯 번째 포도주 잔을 선택하면 총 포도주 양이 33으로 최대로 마실 수 있다.<br>
<br>
**입력**<br>
첫째 줄에 포도주 잔의 개수 n이 주어진다. (1 ≤ n ≤ 10,000) 둘째 줄부터 n+1번째 줄까지 포도주 잔에 들어있는 포도주의 양이 순서대로 주어진다. 포도주의 양은 1,000 이하의 음이 아닌 정수이다.<br>
<br>
**출력**<br>
첫째 줄에 최대로 마실 수 있는 포도주의 양을 출력한다.

```python
# 포도주 잔 n개
n = int(input())
cups = []
for _ in range(n):
    cups.append(int(input()))
#print(cups)
      
d=[0]*10000   # 리스트 범위 벗어난다고 해서 10000 곱해줌
d[0]=cups[0]    #1번 잔 마시는 경우 
d[1]=cups[0]+cups[1]   #1번 잔+2번 잔 마시는 경우  합친 용량
d[2]=max(cups[2]+cups[0], # (3번+1번), 
                cups[2]+cups[1],   # (3번+2번),
                d[1])   # (2번) 중 최대값 용량
# d[i] = i번째(i+1번잔) 마실때 최대가 되는 용량
for i in range(3,n):
    d[i]=max(
        cups[i]+d[i-2],                       # i번째잔만 마셨을 때, i-2까지 최대값 + i 용량
        cups[i]+cups[i-1]+d[i-3],     # i번째잔과 그 전 잔 연속으로 마셨을 때, i-3까지 최대값 + i 용량 + i-1 용량
        d[i-1]        # i번째 잔 안마셨을 때 그 전까지 최대값
        )
print(max(d))
```
