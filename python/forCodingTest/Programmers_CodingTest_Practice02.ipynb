## 제출 답안

```python
def solution(lottos, win_nums):
    x = lottos.count(0)

    lottos_num = []
    for i in lottos:
        if i != 0:
            lottos_num.append(i)

    min_win = 0
    for k in lottos_num:
        if k in win_nums:
            min_win += 1

    max_win = min_win + x

    answer = [0, 0]
    if max_win == 6:
        answer[0] = 1
    elif max_win == 5:
        answer[0] = 2
    elif max_win == 4:
        answer[0] = 3
    elif max_win == 3:
        answer[0] = 4
    elif max_win == 2:
        answer[0] = 5
    else:
        answer[0] = 6

    if min_win == 6:
        answer[1] = 1
    elif min_win == 5:
        answer[1] = 2
    elif min_win == 4:
        answer[1] = 3
    elif min_win == 3:
        answer[1] = 4
    elif min_win == 2:
        answer[1] = 5
    else:
        answer[1] = 6

    return answer
```
