##### No. 5086 배수와 약수
>**문제**<br>
4 × 3 = 12이다. 이 식을 통해 다음과 같은 사실을 알 수 있다. <br>
3은 12의 약수이고, 12는 3의 배수이다.<br>
4도 12의 약수이고, 12는 4의 배수이다.<br>
두 수가 주어졌을 때, 다음 3가지 중 어떤 관계인지 구하는 프로그램을 작성하시오.<br>
첫 번째 숫자가 두 번째 숫자의 약수이다.<br>
첫 번째 숫자가 두 번째 숫자의 배수이다.<br>
첫 번째 숫자가 두 번째 숫자의 약수와 배수 모두 아니다.<br>
<br>
**입력**<br>
입력은 여러 테스트 케이스로 이루어져 있다. 각 테스트 케이스는 10,000이 넘지않는 두 자연수로 이루어져 있다. 마지막 줄에는 0이 2개 주어진다. 두 수가 같은 경우는 없다.<br>
<br>
**출력**<br>
각 테스트 케이스마다 첫 번째 숫자가 두 번째 숫자의 약수라면 factor를, 배수라면 multiple을, 둘 다 아니라면 neither를 출력한다.

```python
while 1:
    a, b = map(int, input().split())
    if a == 0 and b == 0:
        break
    if b%a == 0 :
        print('factor')
    elif a%b == 0:
        print('multiple')
    else:
        print('neither')    
```

##### No. 2981 검문
>**문제** <br>
트럭을 타고 이동하던 상근이는 경찰의 검문을 받게 되었다. 경찰은 상근이가 운반하던 화물을 하나하나 모두 확인할 것이기 때문에, 검문하는데 엄청나게 오랜 시간이 걸린다.<br>
상근이는 시간을 때우기 위해서 수학 게임을 하기로 했다.<br>
먼저 근처에 보이는 숫자 N개를 종이에 적는다. 그 다음, 종이에 적은 수를 M으로 나누었을 때, 나머지가 모두 같게 되는 M을 모두 찾으려고 한다. M은 1보다 커야 한다.<br>
N개의 수가 주어졌을 때, 가능한 M을 모두 찾는 프로그램을 작성하시오.<br>
<br>
**입력**<br>
첫째 줄에 종이에 적은 수의 개수 N이 주어진다. (2 ≤ N ≤ 100)<br>
다음 줄부터 N개 줄에는 종이에 적은 수가 하나씩 주어진다. 이 수는 모두 1보다 크거나 같고, 1,000,000,000보다 작거나 같은 자연수이다. 같은 수가 두 번 이상 주어지지 않는다.<br>
항상 M이 하나 이상 존재하는 경우만 입력으로 주어진다.<br>
<br>
**출력**<br>
첫째 줄에 가능한 M을 공백으로 구분하여 모두 출력한다. 이때, M은 증가하는 순서이어야 한다.

```python
# m으로 나눈 나머지가 모두 r로 같게하는 m 찾기, m > 1
#A = a*m + r
#B = b*m + r
#C = c*m + r

#A-B = (a-b)*m
#B-C = (b-c)*m
# m 은 (a-b), (b-c)의 공약수, 최대공약수 구하고 약수 구하면 됨 (1은 제외)

from math import gcd     # 최대공약수 구하는 함수 gcd(x, y)
from math import sqrt    #제곱근 함수 sqrt(x)

N = int(input())
n = list(int(input()) for _ in range(N))
n.sort()
#print(n)

interval = []
ans = []

for i in range(1, N):
    interval.append(n[i] - n[i - 1])
#print(interval)

prev = interval[0]
for j in range(1, len(interval)):
    prev = gcd(prev, interval[j]) # 최대공약수
#print(prev)

for k in range(2, int(sqrt(prev)) + 1): # 반복 횟수를 제곱근까지만 탐색  # 약수를 2개 조건으로 (몫과 나머지 0인 것) 한번에 탐색하므로
    if prev % k == 0:     # 나머지 0이므로 k는 약수
        ans.append(k)   
        ans.append(prev//k)   # 몫이므로 k는 약수

ans.append(prev)    # 최대공약수도 추가
ans = list(set(ans))   #중복이 있을수 있으니 중복값 제거
ans.sort()
print(*ans)
```

새로 알게된 내용: `print(* )`은 unpacking 해줍니다. →  *[1, 2, 3] 하면 1, 2, 3이 나옴
