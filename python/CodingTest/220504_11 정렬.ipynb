##### No. 2750 수 정렬하기
>**문제**<br>
N개의 수가 주어졌을 때, 이를 오름차순으로 정렬하는 프로그램을 작성하시오.<br>
<br>
**입력**<br>
첫째 줄에 수의 개수 N(1 ≤ N ≤ 1,000)이 주어진다. 둘째 줄부터 N개의 줄에는 수 주어진다. 이 수는 절댓값이 1,000보다 작거나 같은 정수이다. 수는 중복되지 않는다.<br>
<br>
**출력**<br>
첫째 줄부터 N개의 줄에 오름차순으로 정렬한 결과를 한 줄에 하나씩 출력한다.


```python
array = []
N = int(input())
for i in range(N):
    n = int(input())
    array.append(n)
    for j in range(i, 0, -1):
        if array[j] < array[j-1]:
            array[j], array[j-1] = array[j-1], array[j]
        else:
            break

for k in range(len(array)):
    print(array[k])
```

##### No. 2108 통계학

```python
# 시간초과됨
array = []   # (홀수개 = 중앙값 한개라는 뜻)
N = int(input())
for i in range(N):
    n = int(input())
    array.append(n)

#입력된 값 정렬 (리스트)
    for j in range(i, 0, -1):
        if array[j] < array[j-1]:
            array[j], array[j-1] = array[j-1], array[j]
        else:
            break
print(array)

# 산술평균 (소수점 이하 첫째 자리 반올림)   round(실수 ,0)
print(round(sum(array)/len(array)))

# 중앙값
import statistics
print(statistics.median(array))

# 최빈값 (여러개일 경우 두번째 작은 값)
# 리스트의 첫번째 요소 : 첫번째 튜플  mode[0]
# 튜플의 두번째 수 가 max인것 필요  mode[0][1]
# max가 여러개라면 튜플의 첫번째 요소 값중 min 다음 값  mode[0][0]
from collections import Counter
cnt = Counter(array)
mode = cnt.most_common()
#print(mode)

if len(mode) > 1 and mode[0][1]==mode[1][1]: #최빈값 2개 이상
    print(mode[1][0])
else:
    print(mode[0][0])

# 범위 (마지막 요소 값 - 첫번째 요소 값)
print(array[-1] - array[0])
```

```python
# sys로 바꿨는데도 시간초과됨
import sys
array = []
N = int(sys.stdin.readline())
for i in range(N):
    n = int(sys.stdin.readline())
    array.append(n)

    for j in range(i, 0, -1):
        if array[j] < array[j-1]:
            array[j], array[j-1] = array[j-1], array[j]
        else:
            break
print(round(sum(array)/len(array)))

import statistics
print(statistics.median(array))

from collections import Counter
cnt = Counter(array)
mode = cnt.most_common()
if len(mode) > 1 and mode[0][1]==mode[1][1]: 
    print(mode[1][0])
else:
    print(mode[0][0])

print(array[-1] - array[0])
```
