##### No. 15649 N과 M (1)
>**문제**<br>
자연수 N과 M이 주어졌을 때, 아래 조건을 만족하는 길이가 M인 수열을 모두 구하는 프로그램을 작성하시오.<br>
1부터 N까지 자연수 중에서 중복 없이 M개를 고른 수열<br>
<br>
**입력**<br>
첫째 줄에 자연수 N과 M이 주어진다. (1 ≤ M ≤ N ≤ 8)<br>
<br>
**출력**<br>
한 줄에 하나씩 문제의 조건을 만족하는 수열을 출력한다. 중복되는 수열을 여러 번 출력하면 안되며, 각 수열은 공백으로 구분해서 출력해야 한다.  수열은 사전 순으로 증가하는 순서로 출력해야 한다.

```python
from itertools import permutations
N, M = map(int, input().split())
arr = []
for i in range(1, N+1):
    arr.append(i)
#print(arr)

select = list(permutations(arr, M))
#print(select)

for k in select:
#print(k)    #튜플이 나옴 (1, 2, 3, 4)
    for m in k:             # m은 튜플의 요소
        print(m, end = " ")
    print()
```

##### No. 15651 N과 M (3)
>**문제**<br>
자연수 N과 M이 주어졌을 때, 아래 조건을 만족하는 길이가 M인 수열을 모두 구하는 프로그램을 작성하시오.<br>
1부터 N까지 자연수 중에서 M개를 고른 수열<br>
같은 수를 여러 번 골라도 된다.<br>
<br>**입력**<br>
첫째 줄에 자연수 N과 M이 주어진다. (1 ≤ M ≤ N ≤ 7)<br>
<br>**출력**<br>
한 줄에 하나씩 문제의 조건을 만족하는 수열을 출력한다. 중복되는 수열을 여러 번 출력하면 안되며, 각 수열은 공백으로 구분해서 출력해야 한다.<br>
수열은 사전 순으로 증가하는 순서로 출력해야 한다.

```python
from itertools import product
N, M = map(int, input().split())
arr = []
for i in range(1, N+1):
    arr.append(i)
#print(arr)

for j in product(arr, repeat=M):
    #print(j, end=" ")   # 튜플이 나옴
    for m in j:             # m은 튜플의 요소3
        print(m, end = " ")
    print()
```
